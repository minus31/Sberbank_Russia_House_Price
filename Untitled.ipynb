```python
import pandas as pd 

def concatbytime(df1, df2):
    
    """
    이 함수는 "time"과 "value" 가 들어 있는 pandas 데이터 프레임 두개를 인수로 받고, 
    이 두 데이터 프레임을 합치고, "time" 을 기준으로 정렬하는 함수 입니다. 
    ("time" 이 중복되더라도 데이터를 줄이지 않습니다. )
    parameter :: 
    
    df1 : pandas.DataFrame, "time"(datetime.datetime) 과 "value"(float)라는 필드를 포함 하고 있어야 합니다. 
    df2 : pandas.DataFrame, "time" 과 "value"라는 필드를 포함 하고 있어야 합니다. 
    
    return :: 
    
    result : 두 데이터프레임 합쳐지고, 시간 순으로 정렬된 데이터프레임

    """
    
    result = pd.concat([df1, df2], axis=0)
    result = result.sort_values(["time"], ascending=[True])
    result = result.reset_index(drop=True)
    return result 
```
